1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

````python
sns.pointplot(data = prev, x = 'posse_de_veiculo', y = 'renda', dodge = True, errorbar=('ci', 90))
# 'ci' é o nosso intervalo de confiança
````

````python
sns.pointplot(data = prev, x = 'posse_de_imovel', y = 'renda', dodge = True, errorbar=('ci', 90))
````

````python
sns.pointplot(data = prev, x = 'posse_de_veiculo', y = 'renda', hue = 'posse_de_imovel', dodge = True, errorbar=('ci', 90))
````

Analisando os gráficos separadamente, pude observar que tanto a posse de imóvel quanto a posse de veículo  são boas preditoras de renda. A posse de um veículo ou de imóvel contribuem para a indicação de maior renda, isso é reforçado pelos pequenos desvios padrões.
E observando as duas juntas no mesmo gráfico, é possível observar que a posse de imóvel é uma preditora mais precisa se comparada à posse de veículo, isso graças ao seu menor desvio padrão.
