1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python
plt.rc('figure', figsize=(13, 5))
ax = sns.countplot(data = prev, x = 'data_ref', hue = 'posse_de_imovel').tick_params(axis="x", rotation = 90)
````

````python
ax = sns.countplot(data = prev, x = 'data_ref', hue = 'posse_de_veiculo').tick_params(axis="x", rotation = 90)
````

````python
ax = sns.countplot(data = prev, x = 'data_ref', hue = 'estado_civil').tick_params(axis="x", rotation = 90)
````

````python
ax = sns.countplot(data = prev, x = 'data_ref', hue = 'tipo_residencia').tick_params(axis="x", rotation = 90)
````

Analisando os gráficos acima, pude observar uma alteração relevante nos dados da coluna 'educacao', é perceptível um aumento considerável de propostas com superior completo e uma queda significativa de propostas com educação secundária a partir de setembro de 2015. O que merece investigação.
Sobre o restante dos dados qualitativos, eles se mantiveram constantes com o passar do tempo.

2) Para cada variável qualitativa, construa um gráfico mostrando as médias de cada nível da variável no tempo e avalie a estabilidade. Houve alguma alteração importante? Você acha que se pode utilizar essas variáveis em uma modelagem?

````python
ax = sns.pointplot(x = 'data_ref', y = 'posse_de_imovel', data = prev, dodge = True,
                   errorbar=('ci', 95)).tick_params(axis="x", rotation = 90)
````

````python
ax = sns.pointplot(x = 'data_ref', y = 'posse_de_veiculo', data = prev, dodge = True,
                   errorbar=('ci', 95)).tick_params(axis="x", rotation = 90)
````

Analisando os gráficos acima, pude notar pequenas variações da distribuição das médias ao longo dos meses, porém não acredito que causarão problemas na elaboração de modelos de previsão.
