```python
import pandas as pd

import patsy

import numpy as np

import statsmodels.api as sm
```

```python
df = pd.read_csv('previsao_de_renda.csv')
```

#### Como a coluna 'qtd_filhos' tem uma quantidade considerável de valores diferentes, decidi tratá-lo como quantitativa.

1. Ajuste um modelo para prever log(renda) considerando todas as covariáveis disponíveis.
    - Utilizando os recursos do Patsy, coloque as variáveis qualitativas como *dummies*.
    - Mantenha sempre a categoria mais frequente como casela de referência
    - Avalie os parâmetros e veja se parecem fazer sentido prático.  

```python
#Criando matriz de design:
y, X = patsy.dmatrices('np.log(renda) ~ C(sexo) + posse_de_veiculo + posse_de_imovel + qtd_filhos + C(tipo_renda) + C(educacao) + C(estado_civil) + C(tipo_residencia) + idade + tempo_emprego + qt_pessoas_residencia', data=df)
```

```python
#Treinando modelo:
modelo = sm.OLS(y, X).fit()
```

```python
#Exibindo sumário:
modelo.summary()
```

```python
#Definindo as categorias mais frequentes de cada coluna a fim de definir as caselas de referências mais frequente.
print(df.sexo.value_counts())
print()
print(df.tipo_renda.value_counts())
print()
print(df.educacao.value_counts())
print()
print(df.estado_civil.value_counts())
print()
print(df.tipo_residencia.value_counts())
print()
```

```python
#Definindo caselas de referências necessárias.
y, X = patsy.dmatrices('np.log(renda) ~ C(sexo) + posse_de_veiculo + posse_de_imovel + qtd_filhos + C(tipo_renda) + C(educacao, Treatment(2)) + C(estado_civil) + C(tipo_residencia, Treatment(1)) + idade + tempo_emprego + qt_pessoas_residencia', data=df)
modelo = sm.OLS(y, X).fit()
modelo.summary()
```

#### - Todos os parâmetros adicionados aparentam fazer sentido prático.
